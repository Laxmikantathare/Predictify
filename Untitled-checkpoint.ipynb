```python
import numpy as np
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
## Data preprocessing
## Training Image Preprocessing

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip=True)
training_set = train_datagen.flow_from_directory(
      r'F:\\kidney\\kidney\\train',target_size=(224,224),batch_size=32,shuffle=True,class_mode='binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        r'F:\\kidney\\kidney\\test',target_size=(224,224),batch_size=16,shuffle=False,class_mode='binary')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=224 , kernel_size=3 , activation='relu' , input_shape=[224,224,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=224 , kernel_size=3 , activation='relu' ))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2 , strides=2))
```

```python
cnn.add(tf.keras.layers.Dropout(0.5))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=256, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1 , activation='sigmoid'))
```

```python
cnn.compile(optimizer = 'Adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

```python
history=cnn.fit(x = training_set , validation_data = test_set , epochs = 10)
```

```python
loss, accuracy = cnn.evaluate(test_set)
print(f"Test Accuracy: {accuracy}")
```

```python
import pickle
pickle.dump(cnn,open('kidneygraph.pkl','wb'))
```
